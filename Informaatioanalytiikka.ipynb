Datan analysointi aloitetaan tutustumalla luotuun datasettiin ja tekemällä siitä yksinkertaisia havaintoja.
```python
#data preprocessing
import pandas as pd
#produces a prediction model in the form of an ensemble of weak prediction models, typically decisioni
import matplotlib.pyplot as plt
from IPython.display import display

data = pd.read_csv('https://raw.githubusercontent.com/jqzba/FootballAnalytics/main/final_dataset.csv')

data_before_covid = data[:667]
data_covid = data[668:]

n_matches = data_covid.shape[0]
n_features = data.shape[1] - 1
n_homewins = len(data_covid[data_covid.FTR == 'H'])
n_draws = len(data_covid[data_covid.FTR == 'D'])

before_n_matches = data_before_covid.shape[0]
bef_n_homewins = len(data_before_covid[data_before_covid.FTR == 'H'])
bef_draws = len(data_before_covid[data_before_covid.FTR == 'D'])

win_rate = (float(n_homewins) / (n_matches)) * 100
draws = (float(n_draws)/ (n_matches)) * 100

bef_win_rate = (float(bef_n_homewins)/ (before_n_matches) * 100)
bef_draws = (float(bef_draws)/ (before_n_matches)) *100


print("Covid time home winrate: ", win_rate)
print("Before covid home winrate: ", bef_win_rate)

print("Covid draws: ", draws)
print("Before covid draws: ", bef_draws)

print(len(data))

labels = 'HomeWin', 'Draw', 'AwayWin'
sizes = [win_rate, draws, (100 - win_rate - draws)]
before_sizes = [bef_win_rate, bef_draws, (100 - bef_win_rate- bef_draws)]

fig1, (ax1, ax2) = plt.subplots(1,2, figsize=(10,10))
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax1.set_title('Covid time Home/Draw/Away winrates')

ax2.pie(before_sizes, labels=labels, autopct='%1.1f%%', startangle=90)
ax2.set_title("Before covid Home/Draw/Away winrates")

plt.show








```

**PCA- eli pääkomponenttianalyysi**

PCA eli pääkomponenttianalyysi on monimuuttujatekniikka, joka jossa datajoukkoja kuvataan useilla toisiinsa korreloiduilla määrällisillä muuttujilla. Sen tavoitteena on poimia tärkeät tiedot datajoukosta muuttujiksi, joita kutsutaan pääkomponenteiksi ja esittää muuttujien samankaltaisuus erilaisina pistejoukkoina kartoissa/taulukoissa. PCA on todennäköisesti yleisiin tilastollinen menetelmä monimuuttuja joukkuen arviointiin, ja sitä käytetään lähes kaikilla tieteenaloilla. (Abdi et Al. 2010) PCA on pohjimmiltaan uluttovuuden vähentämisalgoritmi, mutta se voi myös olla hyödyllinen työkalu visualisointiin, kohinan suodattamiseen, ominaisuuksien poistoon ja suunnitteluun ja paljon muuta.

Olennainen osa PCA:n käyttöä on arvioida, kuinka monta komponettia tarvitaan tietojen kuvaamiseen. Komponenttien määrää voidaan määrittää varianssiratiolla ja sen kumuloituneen tuloksen suhteella komponenttien lukumäärään, jota lähdetää kokeilemaan ensimmäisenä.


```python

import numpy as np
from sklearn import preprocessing
from sklearn.impute import SimpleImputer as Imputer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA


#remove cols which have tome id or 
cols_to_remove = ['Date', 'HM1', 'AM1', 'HM2', 'AM2', 'HM3',
                  'AM3', 'HM4', 'AM4', 'HM5', 'AM5']
data_test = data.drop(cols_to_remove, axis=1)
data_test.drop(data_test.filter(regex="Unname"),axis=1, inplace=True)


#convert target to numerical values
y=[]
for i in data_test['FTR']:
  if i == 'H':
    y.append(1)
  elif i == 'D':
    y.append(0)
  else:
    y.append(-1)

data_test = data_test.drop('FTR', axis=1)
data_test = pd.get_dummies(data_test)


scaler = preprocessing.MinMaxScaler()
data_rescaled = scaler.fit_transform(data_test)
pca=PCA(n_components=19)
reducedX = pca.fit_transform(data_rescaled)
print("Explained variance of 19 components:\n", pca.explained_variance_ratio_)

PCA_1 = PCA().fit(data_rescaled)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('cumulative explained variance');
plt.show()
```

Explained variance tuottaa vektoreiden pituuksia, joiden pituus on osoitus siitä kuinka "tärkeä" kyseinen akseli on tietojen jakautumisen kuvaamisessa. Kyseissessä datasetissä 3 ominaisuutta erottuvat selvästi muista selittäjistä.

Kumuloituvan selittäjien varianssisuhteesta voidaan havaita, että 2 selittäjää, selittävät enemmän kuin muut, mutta varsinaisesti selittäjien määrässä ei päästä tilanteeseen, että selittäjien määrää olisi kuvaajan perusteella fiksua lähteä vähentämään.

Visualisointi antaa suhteellisen hyvän havainnollistuksen siitä, kuinka paljon ristiin tulokset jalkapallossa menevät ristiin parhaitenkin selittävän muuttujan suhteen. 

Viimeisenä PCA-analyysissä testattiin logistisen regression ennustamistarkkuutta redusoiduilla muutujilla.

```python
pca = PCA(n_components=2)
reducedX = pca.fit_transform(data_test)

fig, ax = plt.subplots(figsize=(20, 21))

colors = ['blue' if l == 1 else('yellow' if l == 0 else 'red')for l in y ]
for dp, c in zip(reducedX, colors):
  ax.scatter(dp[0], dp[1], s=10, color = c)
plt.show()

#Normalize data for dimensionality test
xMean = np.mean(data_test, axis=0)
xDev = np.std(data_test, axis=0)
xNorm = (data_test - xMean) / xDev

x = data_test.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
xMinMax = pd.DataFrame(x_scaled)

#test min_max_scaled data
reducedX = pca.fit_transform(xMinMax)

fig, ax = plt.subplots(figsize=(20, 10))

colors = ['blue' if l == 1 else('yellow' if l == 0 else 'red')for l in y ]
for dp, c in zip(reducedX[:500], colors[:500]):
    ax.scatter(dp[0], dp[1], s=10, color=c)

ax.set_title("MinMaxScaled data")
plt.show()



#predict with reduced dimensionality
for n_comp in range(1, 19):
  pca = PCA(n_components=n_comp)
  reducedX = pca.fit_transform(xNorm)
  X_train, X_test, y_train, y_test = train_test_split(reducedX, y, test_size=0.5, random_state=16)

  clf=LogisticRegression().fit(X_train, y_train)
  score = accuracy_score(y_test, clf.predict(X_test))

  print('Accuracy with {} components: {:.4f}%'.format(n_comp, 100*score))
```


Ennustusten tarkkuudella kasvavilla muuttujamääristä voidaan havaita, että kaikki komponentit eivät erikseen lisättyinä lisää ennustusten tarkkuutta logistisessa regressiossa. Kuitenkin ennustusten tarkkuuksien trendi on kasvava analysoidulla datajoukolla. Tämä mielestäni vahvistaa explained variance metodin luomaa kuvaa siitä, että analyosoidusta datasetistä ei voida havaita selkeästi ylärajaa, jossa ominasuuksien määrällä ei ole enää vaikutusta ennustamiseen. 

PCA analyysin tarkoituksena oli tarkistaa datasetin ominaisuuksien selittävyyttä. Analyysien perusteella ominaisuuksista ei löytynyt selvää linjaa ominaisuuksien määrästä, joka todennäköisesti johtuu siitä, että ominaisuuksia oli dataa toteutettu hyvin rajallisesti. Analyysien perusteella tekisinkin johtopäätöksen, että analysoitavien ominaisuuksien määrää voitaisiin jopa lisätä ennustamistulosten parantamiseksi. 

Jatkossa voisi esimerkiksi kokeilla löytyisikö käytetystä datasetistä ominaisuuksia, joista olisi hyötyä tulosten ennustamisessa. Ainakin yleisömäärään voisi kehitellä mahdollisesti jonkun suhteellisen tekijän, jos yleisö taas joskus pääsee katsomoihin, kun sillä huomattiin olevan huomattavaa vaikutusta otteluihin. Myös esimerkiksi erikoistilanteiden vaikutusta olisi mahdollista kokeilla.

# Jatkokehitys suunnitelmat

Ensimmäisenä jatkokehitys kohteena pitäisin analysoinnissa käytettävän datasetin laajentamista. PCA-analyysien perusteella setistä ei löytynyt selkeästi määrää missä ominaisuuksien lisääminen ei enää suoraan paranna ennustavuutta, vaikka analysoitavat ominaisuudet olivat jopa jonkun verran päällekkäisiä. Datasettiä voisi laajentaa esimerkiksi, käytetystä raakadatasta peliin liittyviä ominaisuuksia, kuten joukkueiden laukaukset (kohti maalia), virhepisteet (booking points), paitsiot, vapaapotkut. Todennäköisesti nämä lisäämällä löytyisi jo selkeästi ominaisuuksia, jotka eivät korreloi lopputuloksen kanssa, mutta alussa mainitsin, että ominaisuuksien valinnasta on vähänlaisesti tutkimuksia, joten ominaisuuksien vaikutukseen olisi hyvä paneutua laskennallisesti. Toisena vaihtoehtona olisi ulkoisten ominaisuuksien parametrisointi, raakadatasta löytyy esim. otteluiden yleisömäärät, joista voisi katsomoiden kapasiteettin verraten luoda keskimääräisen muuttuja, koska visualisoinnissa havaittiin, että yleisöllä on huomattavaa merkitystä kotijoukkueen suoritumiseen. Toinen mahdollisesti parametrsisoitava ominaisuus voisi olla, käytettävien pelaajien parametrisointi, esim. loukkaantuneet pelaajat ovat tiedossa ennen otteluita, joten niiden perusteella olisi jotenkin mahdollista luoda parametri käytettäville pelaajille, mutta tämä vaatisi ulkopuolisen datan tuomista analyysiin. Loput laajentamismahdollisuudet tarkoittaisivatkin datasetin laajentamista uusista datalähteistä, joista voisi saada nykyaikaisia tilastoja kuten xG (maaliodottama), joilla todennäköisesti on korrelaatiota lopputulosten kannalta. Ulkoinen data kuitenkin vaatisi joko datan ostamista, tai data scrabningia, koska kyseisen datan arvo ymmärretään ja sitä ei ole helposti saatavilla ilmaisista lähteistä. 

Toisena jatkokehitys suuntana olisi varmaan muiden koneoppimisen algoritmien testaaminen datasettiin. Tämä ei vaatisi kauheasti vaivaa, mutta todennäköisesti ei myöskään tuottaisi hirveästi. Kandidaatityössäni testatut menetelmät oli havaittu tehokkaimmiksi urheilun tulosten ennustamiseen, joten ennustutarkkuus kehittyisi todennäköisesti tehokkaammin ensimmäisellä kehitysehdotuksella. 

Kolmas kehitysehdotus olisi jonkulaisen tuotteen luominen datasetin ja koneoppimisen perusteella, missä esimerkiksi luotaisiin näkymä tulevista otteluista ja niiden voittajista, koneoppimisen menetelmien tarjoamien todennäköisyyksien mukaan. Nykyisellä datasetillä toteutus olisi mahdollista toteuttaa saman datasetin perusteella, mutta vaatisi huomattavia panostuksia, enkä ole vielä täysin tyytyväinen ominaisuuksien määrään/ ennustustarkkuuteen, joten palaisin taas ensimmäikseksi ensimmäiseen kehitysehodotukseee, koska tämän kehitysehdotuksen toteutus muuttuisi todennäköisesti huomattavasti ensimmäistä kehitysehdoltemaa toteuttamalla.
